```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("Iris.csv")
data.head()
```

# 1.show the count of different types of species

```python
data.groupby(['Species']).size()
```

# 2.find the max,m and mean in values of Sepallength, petallength, sepalwidth and petalwidth

```python
data.describe()
```

# draw a graph:
# 3.scatterplot(for sepallength and petallength)

```python
import seaborn as sns
plt.scatter(data.SepalLengthCm,data.PetalLengthCm)
plt.title('Scatterplot')
plt.xlabel('Sepallength')
plt.ylabel('petallength')
plt.show()
```

# 4.bargraph

```python
objects = ('PetalLengthCm','SepalLengthCm','SepalWidthCm','PetalWidthCm')
plt.bar(objects,len(objects))
plt.show()
```

# 5.jointplot(for sepallength and sepalwidth)

```python
sns.set(style = "white")
g = sns.jointplot(x="SepalLengthCm",y="SepalWidthCm", data=data)
```

# 6.facetgrid(for sepallength and sepalwidth for all species)

```python
g = sns.FacetGrid(data,col = "SepalLengthCm",row = 'SepalWidthCm')
g.map(plt.hist,"Species")
plt.show()
```

# 7.boxplot(for species and sepalwidth)

```python
data.boxplot(column=['SepalWidthCm'],by = "Species")
plt.show()
```

# 8.boxplot & stripplot(for sepallength and all species)

```python
data.boxplot(column=['SepalLengthCm'],by = "Species")
plt.show()
```

```python
sns.stripplot(x=data['SepalLengthCm'],y=data['Species'])
plt.show()
```

# 9.violinplot(for petallength and all species)

```python
sns.violinplot(x = "PetalLengthCm",y = 'Species',hue = "Species",data = data,palete = "muted")
```

# 10.pairplot(for all 4 columns with species)

```python
sns.pairplot(data, hue = "Species")
```
